# Label preprocessing

```python
df['Emotion'].value_counts()
```

```python
df = df[~df['Emotion'].isin(['Calm','Disgust','Surprised'])]
df['Emotion'].value_counts()
```

# Convert emotion labels to numeric

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Emotion'] = label_encoder.fit_transform(df['Emotion'])
# Save the mapping for reference
emotion_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Emotion mapping:", emotion_mapping)
```

```python
print(df['Emotion'].dtype)
```

# Normalization of features

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
# Inspect the data types of all columns
print(df.dtypes)

# Example: Check for specific problematic columns
for col in df.columns:
    print(f"{col}: {df[col].apply(type).unique()}")
```

```python
columns_to_clean = ['Spectral_Centroid', 'Spectral_Rolloff', 'Energy']

# Remove square brackets and convert to numeric for all specified columns
for col in columns_to_clean:
    # Remove square brackets
    df.loc[:,col] = df[col].str.replace(r'[\[\]]', '', regex=True)
    # Convert to numeric
    df.loc[:,col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Store all columns except 'Emotion' in X
X = df.drop(columns=['Emotion'])

# Display X
print(X)
```

```python
y=df['Emotion']
y
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X, y = smote.fit_resample(X, y)
print(pd.Series(y).value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
X=pd.DataFrame(X,columns=df.columns[:-1])
X
```

```python
# Save X and y as CSV files
X.to_csv('X.csv', index=False)
y.to_csv('y.csv', index=False)
```
